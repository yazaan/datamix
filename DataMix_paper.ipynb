```python
%matplotlib notebook
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
```

```python
def plot_elbow_method(train):
  destortions =[]

  for i in range(1,15):
      km = KMeans(n_clusters=i, random_state=1234)
      km.fit(train)
      destortions.append(km.inertia_)


  fig = plt.figure()
  ax = fig.add_subplot(111)
  ax.plot(range(1,15),destortions,marker='o')
  ax.set_xlabel('cluster num')
  ax.set_ylabel('WCSS')
```

```python
plot_elbow_method(train=train)
```
